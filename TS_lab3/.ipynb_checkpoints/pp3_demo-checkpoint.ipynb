## Model

We look at the Stochastic Volatility model

$\alpha_{t} = a \cdot \alpha_{t-1} + s \cdot \eta_t$

$y_{t} = b \exp(- \alpha_t/2) \varepsilon_t$

where $\eta_t$ and $\varepsilon_t$ are two sequences of independet standard Gaussian noise. We also let $\alpha_1$ be a draw from the standard Gaussian distribution.

**E0:** Simulate 250 steps of this model using $a = 0.7$, $s^2 = 0.9$ and $b^2=0.1$

```python
a = 0.7
s2 = 0.9
b2 = 0.1
n = 250
alpha = np.zeros(n)
y = np.zeros(n)
np.random.seed(0)
alpha[0] = np.random.randn(1)
y[0] = np.sqrt(b2)*np.exp(-alpha[0]/2)*np.random.randn(1)
for i in range(n-1):
    alpha[i+1] = a * alpha[i] + np.sqrt(s2)*np.random.randn(1)
    y[i+1] = np.sqrt(b2)*np.exp(-alpha[i+1]/2)*np.random.randn(1)
```

```python
plt.figure()
plt.plot(np.arange(n),y)
plt.plot(np.arange(n),alpha,':')
```

**E1:** For the bootstrap particle filter we need to calculate the weighting function. What will it be in this case?

```python
def logwgtfun(alpha, y, b2):
    # g(y | alpha) = N(y | 0, b2*exp(-alpha))
    #wgt = 1/np.sqrt(2 * np.pi * b2*np.exp(-alpha)) * np.exp(-1/2 * y**2 / (b2 * np.exp(-alpha)))
    logwgt = -1/2 * y**2 / (b2 * np.exp(-alpha)) - 1/2 * np.log(2 * np.pi * b2*np.exp(-alpha))
    return logwgt
```

**E2:** Implement the Bootstrap Particle filter for this data.

$\ell = \sum_{t=1}^{n} \log(1/N \sum_{i=1}^{N} \omega_t^i)$

```python
N = 100
part = np.zeros((n,N))
logwgt = np.zeros((n,N))
wgt = np.zeros((n,N))
llh = 0
part[0,:] = np.random.randn(N)
logwgt[0,:] = logwgtfun(part[0,:],y[0],b2)

for i in range(n-1):
    # Resample
    wgt[i,:] = np.exp(logwgt[i,:])
    wgtsum = np.sum(wgt[i,:])
    wgt[i,:] = wgt[i,:]/wgtsum
    ind = np.random.choice(N, size=N, replace=True, p=wgt[i,:])
    # Propagate
    part[i+1,:] = a * part[i,ind] + np.sqrt(s2) * np.random.randn(N)
    # Weighting
    logwgt[i+1,:] = logwgtfun(part[i+1,:], y[i+1], b2)
    llh = llh + np.log(wgtsum/N)
wgt[-1,:] = np.exp(logwgt[-1,:])
wgtsum = np.sum(wgt[-1,:])
wgt[-1,:] = wgt[-1,:]/wgtsum
llh = llh + np.log(wgtsum/N)
```

```python
plt.figure()
est = np.sum(part*wgt,axis=1)
plt.plot(np.arange(n),est,'k')
plt.plot(np.arange(n),alpha,':')
```
